Variant of UCB algorithm to online-learn a distribution and sample from it at the same time.

Ref for regular UCB: https://arxiv.org/pdf/1402.6028.pdf

```python
import numpy as np
import collections
```

```python
real_distrib = np.asarray([0.1, 1., 2., 10., 2.])
real_normalize = np.sum(real_distrib)
real_distrib /= real_normalize
```

```python
real_distrib
```

```python
N = real_distrib.shape[0]
samples = [ [] for _ in range(N) ]
T = 1
```

```python
def current_mean_and_std():
    global samples
    mean = [ np.average(x) for x in samples ]
    mean = [ np.nan_to_num(x) for x in mean ]
    std = [np.std(x) for x in samples]
    std = [(1.e6 if len(l)<=1 else x/np.sqrt(len(l))) for l, x in zip(samples, std) ]
    return np.asarray(mean), np.asarray(std)
```

```python
def compute_sampling_probs():
    global T
    global samples
    m , _ = current_mean_and_std()
    c = np.asarray([len(s) for s in samples])
    c = np.sqrt(2*np.log(T)/c)
    mask = np.isinf(c) | np.isnan(c)
    c[mask] = 1.e6
    w = np.abs(m) + 10.*c
    w /= np.sum(w)
    return w
```

```python
def add_samples(p, M):
    global samples
    for i in np.random.choice(len(samples), M, p = p):
        x = np.random.normal(real_distrib[i], 0.3)
        samples[i].append(x)
```
