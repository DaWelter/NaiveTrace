MIS Combination of Camera and Image Samples
===========================================
When rendering an image with path tracing one can start from pixels and record the computed path
weight. But one can also start from lights and and see what pixels the light path hit. Depending
the definition of the function We [Veach's Thesis] sensing areas of pixels can even overlap.

This notebook is an extremely idealized numerical experiment on how to combine both kinds of samples
via MIS.

```python
import numpy as np
import matplotlib.pyplot as pyplot
import scipy.stats
%matplotlib inline
pyplot.ioff()
```

```python
class ImageBuffer(object):
    def __init__(self, size):
        self.sum = np.zeros(size)
        self.sum2 = np.zeros(size)
    
    def add(self, idx, other):
        self.sum[idx] += other
        self.sum2[idx] += np.square(other)

    def avg_std(self, num_samples):
        avg = self.sum/num_samples
        avg2 = self.sum2/num_samples
        tmp = avg2 - avg*avg
        std = np.sqrt(tmp)
        std /= np.sqrt(num_samples) # standard deviation of the mean
        return avg, std
```

```python
# Mis weight. Power heuristic. mis_weight(q1,q2)+mis_weight(q2,q1)=1!
def mis_weight(q1, q2):
    n = 1.
    return q1**n / (q1**n + q2**n)
```

```python
def setup_problem(sensor_width, object_width, object_offset=0.):
    # There is an array of NSENSORS number of sensors with some area in the [0,1] interval where they respond to an input.
    NSENSORS = 10
    SENSOR_CENTERS = (0.5+np.arange(NSENSORS)) / NSENSORS
    DX = SENSOR_CENTERS[1]-SENSOR_CENTERS[0]
    SENSOR_WIDTH=sensor_width
    OBJECT_WIDTH=object_width
    class Sensor(object):
        def __init__(self, i):
            self.density = scipy.stats.norm(loc=SENSOR_CENTERS[i], scale=DX*SENSOR_WIDTH) # Gaussian centered at SENSOR_CENTERS[i]
            self.pdf = self.density.pdf
            self.rvs = self.density.rvs # Returns sample according to pdf.
            self.loc = SENSOR_CENTERS[i]
        def We(self, x):
            # Some arbitrary multiple of the pdf.
            return self.pdf(x) * 0.42
    # The 'object' being imaged. It is really just a box function.
    class Object(object):
        def __init__(self):
            self.density = scipy.stats.uniform(loc=0.5-0.5*OBJECT_WIDTH+object_offset, scale = OBJECT_WIDTH)
            self.pdf = self.density.pdf
            self.rvs = self.density.rvs
        def Le(self, x):
            # Some arbitrary multiple of the pdf.
            return self.pdf(x) * 2.
    sensors = map(Sensor, range(NSENSORS))
    imaged = Object()
    return sensors, imaged
```

```python
def compute_ref_and_visualize(SENSOR_CENTERS, sensors, imaged):
    NSENSORS = len(sensors)
    X = np.linspace(0., 1., 1000)
    imaged_rasterized = map(imaged.Le, X)
    # Compute reference image by taking ...
    reference_image = np.empty(NSENSORS)
    for i in range(NSENSORS):
        # ... the inner product of sensor rensponse with image.
        reference_image[i] = np.sum(imaged_rasterized * np.apply_along_axis(sensors[i].We, 0, X))*(X[1]-X[0])
    # Visualization.
    pyplot.plot(X, imaged_rasterized)
    for i in range(NSENSORS):
        pyplot.plot(X, map(sensors[i].We, X), c='r')
    pyplot.plot(SENSOR_CENTERS, reference_image, c='k')
    pyplot.show()
    return reference_image
```

```python
def display_results(x, stuff, reference_image):
    (sens_avg, sens_std), (im_avg, im_std), (total_avg, total_std) = stuff
    pyplot.errorbar(x, sens_avg, yerr = sens_std, c = 'r', ls = '--', label = 'sens')
    pyplot.errorbar(x, im_avg, yerr = im_std, c = 'r', ls = '-.', label = 'im')
    pyplot.errorbar(x, total_avg, yerr = total_std, c = 'm', label = 'mis')
    pyplot.plot(x, reference_image, c = 'k', ls = ':', label = 'ref')
    pyplot.legend()
    pyplot.show()
```

```python
def OriginalWayToSampleImageWithOverlappingSensorSupport(sensors, imaged):
    NITERS = 10
    NSENSORS=len(sensors)
    image = ImageBuffer(NSENSORS)  # By sampling the sensors.
    image2 = ImageBuffer(NSENSORS) # By sampling the image.
    total = ImageBuffer(NSENSORS)  # Combination.
    NUM_COMBINED_SAMPLES = (1+NSENSORS)*NITERS
    WEIGHT_NUM_SENSE = float(NUM_COMBINED_SAMPLES)/NITERS
    WEIGHT_NUM_OBJ   = float(NUM_COMBINED_SAMPLES)/(NITERS*NSENSORS)
    for iter in xrange(NITERS):
        for i in xrange(NSENSORS):
            # First take a sample from sensor i and record its response which gives the pixel value.
            x = sensors[i].rvs()
            spdf = sensors[i].pdf(x)
            ipdf = imaged.pdf(x)
            pixelval = sensors[i].We(x)/spdf*imaged.Le(x)
            image.add(i, pixelval)
            total.add(i, pixelval*WEIGHT_NUM_SENSE*mis_weight(spdf, NSENSORS*ipdf))
            # Now for every sensor sample I also take a light sample. But here it is not known a-priori 
            # which sensor will response. In fact multiply sensors might respond. In effect, there are
            # NITERS*NSENSORS object samples for which the response of each sensor must be considered.
            # The background for this configuration is that in BDPT one shoots one light path per pixel and
            # that light path can be connected to the camera vertex resulting in potentially multiple
            # responding pixel.
            x = imaged.rvs()
            y = imaged.Le(x)
            ipdf = imaged.pdf(x)
            for j in xrange(NSENSORS):
                spdf = sensors[j].pdf(x)
                pixelval = sensors[j].We(x)*y/ipdf
                image2.add(j, pixelval)
                total.add(j, mis_weight(NSENSORS*ipdf, spdf)*WEIGHT_NUM_OBJ*pixelval)
    return image.avg_std(NITERS), image2.avg_std(NITERS*NSENSORS), total.avg_std(NUM_COMBINED_SAMPLES)
```

Re: WEIGHT_NUM_SENSE and WEIGHT_NUM_OBJ

The total MIS weighted estimator has the form:
$$F = \sum_{i=1}^n w_{a,i} \frac{1}{n} A_i + \sum_{i=1}^m w_{b,i} \frac{1}{m} B_i $$
In order to calc statistics I want to bring it into the form:
$$F = \frac{1}{m+n} \sum_{i=1}^{n+m} C_i$$
Hence I reformulate F to read
$$F = \frac{1}{m+n} \left( \sum_{i=1}^n w_{a,i} \frac{m+n}{n} A_i + \sum_{i=1}^m w_{b,i} \frac{m+n}{m} B_i \right)$$
Thus
$$\text{WEIGHT_NUM_SENSE} = \frac{m+n}{n}, \text{WEIGHT_NUM_OBJ} = \frac{m+n}{m}$$

So, overlapping sensor responses. But can I simplify the computations by removing all but one response term. Russian Roulette comes to mind. 

```python
def EstimatorUsingRussianRoulette(sensors, imaged):
    def compute_sensor_survival_probabilities(x):
        sensor_selection_prob = np.asarray([ abs(s.We(x)) for s in sensors ])
        sensor_selection_prob /= np.sum(sensor_selection_prob)
        return sensor_selection_prob

    def compute_sensor_survival_probability(i, x):
        return compute_sensor_survival_probabilities(x)[i]

    def select_surviving_sensor(x):
        p = compute_sensor_survival_probabilities(x)
        i, = np.random.choice(np.arange(len(sensors)), size=1, p=p)
        return i, p[i]

    NITERS = 10
    NSENSORS = len(sensors)
    image = ImageBuffer(NSENSORS)  # By sampling the sensors.
    image2 = ImageBuffer(NSENSORS) # By sampling the image.
    total = ImageBuffer(NSENSORS)  # Combination.
    NUM_COMBINED_SAMPLES = (1+NSENSORS)*NITERS
    WEIGHT_NUM_SENSE = float(NUM_COMBINED_SAMPLES)/NITERS
    WEIGHT_NUM_OBJ   = float(NUM_COMBINED_SAMPLES)/(NITERS*NSENSORS)
    for iter in xrange(NITERS):
        for i in xrange(NSENSORS):
            # First take a sample from sensor i and record its response which gives the pixel value.
            x = sensors[i].rvs()
            spdf = sensors[i].pdf(x)
            ipdf = imaged.pdf(x)*compute_sensor_survival_probability(i, x)
            pixelval = sensors[i].We(x)/spdf*imaged.Le(x)
            image.add(i, pixelval)
            total.add(i, pixelval*WEIGHT_NUM_SENSE*mis_weight(spdf, NSENSORS*ipdf))
            # Now for every sensor sample I also take a light sample.
            x = imaged.rvs()
            y = imaged.Le(x)
            # The following is kinda like in OriginalWayToSampleImageWithOverlappingSensorSupport
            # but I remove all but one response probabilistically via russian roulette, picking
            # the surviving sensor in proportion to the response function (We).
            j, select_prob = select_surviving_sensor(x)
            ipdf = imaged.pdf(x)*select_prob
            spdf = sensors[j].pdf(x)
            pixelval = sensors[j].We(x)*y/ipdf
            image2.add(j, pixelval)
            total.add(j, mis_weight(NSENSORS*ipdf, spdf)*WEIGHT_NUM_OBJ*pixelval)
    return image.avg_std(NITERS), image2.avg_std(NITERS*NSENSORS), total.avg_std(NUM_COMBINED_SAMPLES)
```

```python
def run(*args, **kwargs):
    sensors, imaged = setup_problem(*args, **kwargs)
    x = [s.loc for s in sensors]
    reference = compute_ref_and_visualize(x, sensors, imaged)
    stuff = OriginalWayToSampleImageWithOverlappingSensorSupport(sensors, imaged)
    display_results(x, stuff, reference)
    stuff = EstimatorUsingRussianRoulette(sensors, imaged)
    display_results(x, stuff, reference)
```

```python
run(sensor_width=0.2, object_width=0.4)
```

```python
run(0.5, 0.05)
```
