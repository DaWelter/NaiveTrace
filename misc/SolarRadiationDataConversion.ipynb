```python
import matplotlib.pyplot as pyplot
import os, sys
import numpy as np
import json
```

```python
with open('solar_spectrum_E490_AM0_dataset.dat', 'r') as f:
    lines = f.readlines()
```

```python
def map_line(l):
    l = map(str.strip, l.split())
    l = map(float, l)
    return l
data = map(map_line, lines)
data = np.asarray(data)
```

```python
mask = np.bitwise_and(data[:,0]>0.300, data[:,0]<0.800)
spectrum = data[mask,:]
spectrum[:,0] *= 1000.
```

```python
%matplotlib inline
```

```python
pyplot.plot(spectrum[:,0], spectrum[:,1])
```

```python
NBINS = 36
DLAMBDA = 10. # width of the bins
bin_boundaries = 370. + np.arange(NBINS+1)*DLAMBDA
print bin_boundaries
bin_centers = np.average([bin_boundaries[1:], bin_boundaries[:-1]], axis=0)
print bin_centers
```

```python
downsampled, _ = np.histogram(spectrum[:, 0], bins = bin_boundaries, weights = spectrum[:,1])
counts,_       = np.histogram(spectrum[:, 0], bins = bin_boundaries)
print downsampled
print counts
downsampled /= counts
```

```python
pyplot.ioff()
pyplot.bar(bin_boundaries[:-1], downsampled, bin_boundaries[1:]-bin_boundaries[:-1])
pyplot.plot(spectrum[:,0], spectrum[:,1], color = 'r')
pyplot.draw()
```
