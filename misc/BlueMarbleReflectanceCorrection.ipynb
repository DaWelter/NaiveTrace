Blue Marble Reflectance Correction
----------------------------------

The Blue Marble data set is a set of texture images of the earth.
https://earthobservatory.nasa.gov/Features/BlueMarble/

It comprises a reflectance map. However, its rgb values produced by applying a cubic mapping function to real reflectance values. See [1]. I attempt to correct for this mapping and output a gamma corrected (non-linear) RGB image.

Refs:
* [1] https://earthobservatory.nasa.gov/Features/BlueMarble/bmng.pdf, Sec 2.1 and 2.5.
* [2] MODIS Surface Reflectance User’s Guide (Section 3.1.5) http://modis-sr.ltdri.org/guide/MOD09_UserGuide_v1.4.pdf

```python
def g(x, *p):
    x = x/255.
    #return x*p[0] + x*x*p[1] + x*x*x*p[2]
    # 2*x*p[1] + 3x*x*p[2] + p[0] = c
    #c = 0.1
    #p0 = c -2.*x*p[0] -3*x*x*p[1]
    #return x*p0 + x*x*p[0] + x*x*x*p[1]
    return np.power(x, p[0])*p[1] + 0.3*x
p, _ = scipy.optimize.curve_fit(
    g,
    rgb,
    refl,
    p0 = (1., 1.))
print p
pyplot.plot(rgb, refl)
pyplot.plot(x, g(x,*p))
pyplot.show()
def bmng_value_to_reflectance(x):
    return g(x, *p)
```

```python
def LinearToSRGB(x):
    mask_small = x.ravel() < 0.0031308
    y = np.empty_like(x)
    y.ravel()[mask_small] = 0.0031308*x.ravel()[mask_small]
    mask_large = ~mask_small
    y.ravel()[mask_large] = 1.055*np.power(x.ravel()[mask_large], 1.0/2.4) - 0.055
    return y
```

Color Correction
================

My renderer takes in images in sRGB color space and converts them to higher resolved discrete spectra. What follows is an attempt to convert BMNG Topography maps to sRGB images suitable for my renderer.

I do this by first constructing a discrete spectrum from the RGB values and projecting it to XYZ by taking the inner product with XYZ color matching function. Then I multiply by the XYZ to sRGB transform matrix.

```python
# My default spectral discretization.
NBINS = 36
DLAMBDA = 10. # width of the bins
bin_boundaries = 370. + np.arange(NBINS+1)*DLAMBDA
bin_centers = np.average([bin_boundaries[1:], bin_boundaries[:-1]], axis=0)
```

```python
# Define spectral response curves of the MODIS sensors. 
# Take box filters for the spectral ranges given in the MODIS database manual [2].
# Normalize so that the inner product of the response with a flat spectrum becomes 1.
def between(x, a, b):
  return np.logical_and(x <= b, x > a)  
sensor_response_curves = np.zeros((3,NBINS), np.float32)
sensor_response_curves[0,between(bin_centers, 620,670)] = 1.
sensor_response_curves[1,between(bin_centers, 545,565)] = 1.
sensor_response_curves[2,between(bin_centers, 459,479)] = 1.
sensor_response_curves /= np.sum(sensor_response_curves, axis=1)[:,np.newaxis]
pyplot.plot(bin_centers, sensor_response_curves[0], c='r')
pyplot.plot(bin_centers, sensor_response_curves[1], c='g')
pyplot.plot(bin_centers, sensor_response_curves[2], c='b')
pyplot.gca().set(ylim=(-0.1, 1.1))
pyplot.show()
```

```python
# Build a spectral bases, one base per sensed frequency band.
# I want the spectra to be relatively smooth and more importantly,
# the spectrum i corresponding to band i to give a large response for band i
# and low response for the other bands.
def response_matrix(spectra):
    # Compute matrix of responses for pairwise combinations of sensor band and spectrum.
    rgb_response = np.asarray([(spectra[i]*sensor_response_curves) for i in xrange(3)])
    rgb_response = np.sum(rgb_response, axis=2)
    return rgb_response
def constraint_func1(spectra):  # Spectra need to be positive.
    return spectra
def constraint_func2(spectra):  # Must sum up to one for each wavelength.
   return np.sum(spectra.reshape(3, NBINS), axis=0)-1.
def constraint_func3(spectra):  # Need to be less than 1
    return 1.-spectra
def objective_func(spectra):
    spectra = spectra.reshape(3, NBINS)
    rgb_response = response_matrix(spectra)
    response_that_shall_be_one = np.asarray([rgb_response[i,i] for i in xrange(3)])
    response_that_shall_be_zero = np.asarray(
        [rgb_response[i,j] for (i,j) in np.ndindex(3,3) if (i!=j)])
    cost1 = np.average(np.square(response_that_shall_be_one-1.))
    cost2 = np.average(np.square(response_that_shall_be_zero))
    # In addition to the 'sensitivity' related cost, add some smoothness regularization.
    # If possible I want to drive the sum of the spectra to a flat 
    dx = spectra[:,1:] - spectra[:,:-1]
    ddx = dx[:,1:] - dx[:,:-1]
    soft_sum_constraint = np.average(np.square(constraint_func2(spectra)))
    #soft_sum_constraint = 1.
    reg1 = np.average(np.square(dx))
    reg2 = np.average(np.square(ddx))
    return cost1 + 50.* cost2 + 5. * reg1 + 10. * reg2 + 1. * soft_sum_constraint
result = scipy.optimize.minimize(
    objective_func, 
    np.random.normal(1., 0.1, size=(3,NBINS)),
    constraints=[
        {'type' : 'ineq', 'fun' : constraint_func1},
        {'type' : 'ineq', 'fun' : constraint_func3},
        #{'type' : 'eq', 'fun' : constraint_func2}
    ],
    options = {'disp' : True, 'eps' : 1.e-5})
spectral_basis = result.x.reshape(3, NBINS)
pyplot.plot(bin_centers, spectral_basis[0], c='r')
pyplot.plot(bin_centers, spectral_basis[1], c='g')
pyplot.plot(bin_centers, spectral_basis[2], c='b')
pyplot.plot(bin_centers, np.sum(spectral_basis,axis=0), c='k')
pyplot.plot(bin_centers, sensor_response_curves[0], c='r', ls=':')
pyplot.plot(bin_centers, sensor_response_curves[1], c='g', ls=':')
pyplot.plot(bin_centers, sensor_response_curves[2], c='b', ls=':')
pyplot.gca().set(ylim=(-0.1, 1.1))
pyplot.show()
for (i,j) in np.ndindex(3,3):
    print "Sensor %i reponse to base %i = %0.3f" % (j, i, response_matrix(spectral_basis)[i,j])
```

```python
# Load XYZ response curves.
# Data from http://cvrl.ioo.ucl.ac.uk/cmfs.htm
# CIE 1931 2-deg, XYZ CMFs modified by Judd (1951) and Vos (1978)
import csv
with open('ciexyzjv.csv', 'rb') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    data = []
    for row in reader:
        data.append(map(float, row))
data = np.asarray(data).T
# Interpolate to my binned spectra
xyz_response_curves = scipy.interpolate.interp1d(data[0,:], data[[1,2,3],:], kind='quadratic', bounds_error=False, axis=1, fill_value=data[[1,2,3],0])(bin_centers)
# Do they look okay?
pyplot.plot(bin_centers, xyz_response_curves[0], color='r')
pyplot.plot(bin_centers, xyz_response_curves[1], color='g')
pyplot.plot(bin_centers, xyz_response_curves[2], color='b')
pyplot.show()
```
